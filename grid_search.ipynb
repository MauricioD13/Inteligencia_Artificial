```python
df = pd.read_csv(r'CarPrice_Assignment.csv')
```

```python

data = pd.DataFrame(df, columns=['horsepower', 'price'])
features = np.array(data['horsepower'])
labels = np.array(data['price'])

plt.scatter(features,labels)
plt.show()
```

```python
"""Parametros -> o0 y o1
   labels (Etiquetas) -> resultados 
   characteristics -> entradas
"""
def error_len(labels, features):
    if len(labels) != len(features):
        raise ValueError("Length Error")
    
def costo(parameters,labels,features,n):
    
    aux = parameters[0] + parameters[1]*features
    return (1/(2*n))*np.sum(np.power((aux - labels),2))  
```

```python
error_len(labels, features)
n = len(labels)
for parameters[0] in np.arange(10,15,0.01):
    for parameters[1] in np.arange(10,12,0.01):
        
        results[float(costo(parameters, labels, features,n))] = (parameters[0], parameters[1])
        
minimum = min(results)
print(f'Resultado: {minimum}, parametros:{results[minimum]}')
#print(results)
```
