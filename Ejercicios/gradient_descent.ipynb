# Descenso de gradiente

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'CarPrice_Assignment.csv')
data = pd.DataFrame(df, columns=['horsepower', 'price'])
features = np.array(data['horsepower'])
labels = np.array(data['price'])
parameters = [10000,10000]
_parameters = [0,0]
```

```python
"""Parametros -> o0 y o1
   labels (Etiquetas) -> resultados 
   characteristics -> entradas
"""
def error_len(labels, features):
    if len(labels) != len(features):
        raise ValueError("Length Error")
    
def costo(parameters,labels,features,n):
    
    aux = parameters[0] + parameters[1]*features
    return (1/(2*n))*np.sum(np.power((aux - labels),2))

def gradiente_par0(parameters,labels,features,n,):
    aux = parameters[0] + parameters[1]*features
    return (1/n) * np.sum((aux - labels))
def gradiente_par1(parameters,labels,features,n,):
    aux = parameters[0] + parameters[1]*features
    return (1/n) *(np.sum((aux - labels) * features))
```

```python
n= len(labels)
parameters = [100,50000]
alpha = 0.00001
par = []
plt.plot(labels)
for i in range(10):
    plt.plot(_parameters[0], _parameters[1],'o',c='r')
    
    _parameters[0] = parameters[0] - alpha * gradiente_par0(parameters, labels, features, n)
    _parameters[1] = parameters[1] - alpha * gradiente_par1(parameters, labels, features, n)
    parameters = _parameters  
    print(_parameters)

plt.show
```
