# Descenso de gradiente

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
""" Tomando los valores de caballos de fuerza y el precio de una dataset de carros se busca hacer una regresion lineal
    Caracteristicas serian los caballos de fuerza 
    Etiquetas seria el precio
"""
df = pd.read_csv(r'CarPrice_Assignment.csv')
data = pd.DataFrame(df, columns=['horsepower', 'price'])
features = np.array(data['horsepower'])
labels = np.array(data['price'])
```

```python
"""Parametros -> o0 y o1
   labels (Etiquetas) -> resultados 
   characteristics -> entradas
"""
"""
 Función de costo que el entran los dos thetas, X y Y
"""
def cost(theta0,theta1,x,y):
  #donde theta0 y theta1 son los parámetros a probar,
  #x y y son nparrays de la misma dimensión
  n=x.size
  #aqui podrian depurar que los tamaños de los arreglos sean iguales

  #calculo del costo en la variable J
  J=(1/(2*n))*np.sum(np.power(( (theta0+theta1*x)-(y) ),2)) #terminen ustedes... 
  time.sleep(1)
  return J
"""
 Función de theta0 que el entran los dos thetas, X y Y
"""
def gradient_theta0(theta0,theta1,x,y):
  n=x.size# para normalizar el número de muestras
  return (1/n)*np.sum(( (theta0+theta1*x)-(y) ))
"""
 Función de theta1 que el entran los dos thetas, X y Y
"""
def gradient_theta1(theta0,theta1,x,y):
  n=x.size# para normalizar el número de muestras
  return (1/n)*np.sum( ((theta0+theta1*x)-(y))*x )

def gradient_descent(alpha,theta0,theta1,x,y):
  G0=gradient_theta0(theta0,theta1,x,y)
  G1=gradient_theta1(theta0,theta1,x,y)
  
  theta0km1=theta0-alpha*G0
  theta1km1=theta1-alpha*G1
  return theta0km1,theta1km1
```

```python
n= len(labels)
theta0 = 0.001
theta1 = 0.5
alpha = 0.00001
for i in range(100):
    theta0, theta1 = gradient_descent(alpha,theta0,theta1,features,labels)
    plt.plot(features, theta0+theta1*features,c='r')
plt.scatter(features,labels)
plt.plot(features, theta0+theta1*features)
```
